# Funci칩n $\dot{x} = r + x^2$

```
# An치lisis de la funci칩n



x = np.linspace(-2,2,100)

def func(x, r):
  x_dot = r + x**2
  return x_dot

# r < 0
r1 = -2
x1 = func(x, r1)
plt.plot(x,x1, label='r < 0')

# r = 0
r2 = 0
x2 = func(x, r2)
plt.plot(x,x2, label='r = 0')

# r > 0
r3 = 2
x3 = func(x, r3)
plt.plot(x,x3, label='r > 0')

plt.grid()
plt.legend(loc='best')
plt.show()
```
