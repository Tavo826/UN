```
#Sección 3.4 pag 55  Nonlinear dynamics and chaos S. Strogatz
import numpy as np
import matplotlib.pyplot as plt

# Equation in exercise_2_2_1

x = np.arange(-4,4,.1)
r = 4
dx = r*x - x**3


fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.spines['left'].set_position(('data',0.0))
ax.spines['bottom'].set_position(('data',0.0))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
#ax.set_xticks([-4,-2,2,4])
#ax.set_yticks([-20,20,40,60])

plt.plot(x,dx)
plt.show()
```

```
from scipy.integrate import solve_ivp
# Sección 3.4 pag 55  Nonlinear dynamics and chaos S. Strogatz
def dx(t, x, r, b): return r*x - (x**2)-d

a,b = 4,16
x0 = np.array([-6, -4, -3, -2,-1,0, 1, 1.5, 1.8, 2, 2.0001])
tspan = [0, 0.6]
sol = solve_ivp(dx, tspan, x0,args=(a,b),\
                dense_output = True)

#print(sol.t)
#print(sol.y[0])
t = np.linspace(tspan[0],tspan[1],101)
x = sol.sol(t)
for i in range(x0.size):
  #plt.plot(sol.t,sol.y[i])
  plt.plot(t,x[i])
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
#Logistic Equation N_dot = r*N*(1-N/k)
N = np.arange(0,12,.1)
r,k = 1,10
dN = r*N*(1-N/k)


fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.spines['left'].set_position(('data',0.0))
ax.spines['bottom'].set_position(('data',0.0))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.set_xticks([5,10])
ax.set_yticks([0,2])

plt.plot(N,dN)
plt.show()
```

```
from scipy.integrate import solve_ivp
# Equation in exercise_2_2_1
def dN(t, N, r, k): 
  return  r*N*(1-N/k)

r,k = 1,10
N0 = np.arange(2,20,2)
tspan = [0, 10]
sol = solve_ivp(dN, tspan, N0, args = (r,k),\
                dense_output = True)

#print(sol.t)
#print(sol.y[0])
t = np.linspace(tspan[0],tspan[1],101)
x = sol.sol(t)
for i in range(N0.size):
  #plt.plot(sol.t,sol.y[i])
  plt.plot(t,x[i])
plt.show()
```
